```python
# Creating a Back Testing System for LOOP


def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):   # every year has almost 250 trading days therefore, 250*10 = 2500 which means a decade long data
    all_predictions = []

    for i in range(start, data.shape[0], step):
      train = data.iloc[0:i].copy()
      test = data.iloc[i:(i+step)].copy()
      predictions = predict(train, test, predictors, model)
      all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
# still our accuracy is poor

# therefore we would use rolling averages

horizons = [2,5,60,250,1000]
new_predictors = []

for horizion in horizons:
  rolling_averages = sp500.rolling(horizion).mean()

  ratio_column = f"Close_Ratio_{horizion}"
  sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
  new_predictors.append(ratio_column)

  trend_column = f"Trend_{horizion}"
  sp500[trend_column] = sp500.shift(1).rolling(horizion).sum()["Target"]

  new_predictors += [ratio_column, trend_column]
```

```python
# Update our model


model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# Adjusting the Predict function slightly

def predict(train,test,predictors,model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= .6] = 1 # Model Go UP
  preds[preds < .6] = 0  # Model GO DOWN
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis = 1)
  return combined
```

```python
# With this what we trying to do is have choosing new predictors whereas Close, High etc means nothing
# hence we select ratios

predictions = backtest(sp500, model, new_predictors)
```
